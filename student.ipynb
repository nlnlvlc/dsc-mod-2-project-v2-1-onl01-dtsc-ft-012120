```python
# Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

df = pd.read_csv('kc_house_data.csv')
```

```python
# Checking df's info, types, and looking for missing data
df.info()
```

```python
#Checking for null values
df.isna().sum()
```

```python
#check values for waterfront to handle null values
df['waterfront'].value_counts(normalize=True)
```

```python
#check values for view to handle null values
df['view'].value_counts(normalize=True)
```

```python
#check values for yr_renovated to handle null values
df['yr_renovated'].value_counts(normalize=True)
```

```python
#Percentage of values in each series is '0' for >90%
#filling in null values to 0
with_null = ['waterfront', 'view', 'yr_renovated']
for cols in df[with_null]:
    df[cols].fillna(value=0, inplace=True)
```

```python
df.info()
```

```python
#check series with object types: date
df['date'].head()
df['date'].value_counts(normalize=True)
```

```python
#check series with object types: sqft_basement
df['sqft_basement'].head()
```

```python
#convert date into datetime
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y', errors='coerce')
```

```python
df.info()
```

```python
#most value types in sqft_basement seem to be float64
#check for object types

df['sqft_basement'].value_counts(normalize=True)
```

```python
#remove rows with '?' values
df.drop(df.index[df['sqft_basement'] == '?'], inplace = True)
```

```python
#check that rows were successfully removed across df
df.info()
```

```python
#change type from object to float64(
df['sqft_basement'] = df['sqft_basement'].apply(pd.to_numeric)
```

```python
#check to make sure object types were successfully changed
df.info()
```

```python
#create columns for months
df['month'] = df['date'].dt.month
```

```python
#assign names to months

month_name = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

#replace current data with month_name
df['month'] = df['month'].replace(month_name)
```

```python
#create a column for years
df['year'] = df['date'].dt.year
```
